# Статистика выживания по всем классам:

# Самое популярное женское имя на корабле:

```python
df = pd.read_csv('train.csv')
index = 0
print len(df)
t = len(df)
while index < t:
    s2 = df.iget_value(index,3)
    s3 = s2.split()
    index3 = 0
    f = False
    while index3<len(s3) and f==False:
        if s3[index3][0] == '(':
            f = True
        else:
            index3 = index3 + 1
    if f == True:
        s2 = s3[index3]
    else:
        index3 = 0
        while index3<len(s3) and f==False:
            if s3[index3][len(s3[index3])-1] == '.':
                f = True
            else:
                index3 = index3 + 1
        s2 = s3[index3+1]
    s2 = s2.replace('(','')
    s2 = s2.replace(')','')
    df.set_value(index,'Name(simple)',s2)
    index = index+1
df2 = df[df['Sex']=='female']
df2['Name(simple)'].value_counts()[0:10].plot(kind = 'bar' , title='Top 10 female names on the ship')
print 'The most popular name is '+str(df2['Name(simple)'].mode()[0])+'. There were '+str(df2['Name(simple)'].value_counts()[0])+' of them!'
```

# Связь между номером билета и классом пассажира

```python
df = pd.read_csv('train.csv')

df2 = df[df['Pclass']==1]
df2 = df2['Ticket'].dropna()
print len(df2)
tickets1 = df2.get_values()
for index in range(0,len(tickets1)):
    if len(tickets1[index].split())>1:
        tickets1[index] = tickets1[index].split()[1][0]
    else:
        tickets1[index] = tickets1[index][0]
tickets1 = sort(tickets1)

df2 = df[df['Pclass']==2]
df2 = df2['Ticket'].dropna()
print len(df2)
tickets2 = df2.get_values()
for index in range(0,len(tickets2)):
    if len(tickets2[index].split())>1:
        tickets2[index] = tickets2[index].split()[1][0]
    else:
        tickets2[index] = tickets2[index][0]
tickets2 = sort(tickets2)

df2 = df[df['Pclass']==3]
df2 = df2['Ticket'].dropna()
print len(df2)
tickets3 = df2.get_values()
for index in range(0,len(tickets3)):
    if len(tickets3[index].split())>1:
        tickets3[index] = tickets3[index].split()[1][0]
    else:
        tickets3[index] = tickets3[index][0]
tickets3 = sort(tickets3)

df1 = pd.DataFrame(tickets1)
df1.drop_duplicates()
for index in range(0,len(df1)):
    df1.set_value(index,'Class',1)
df2 = pd.DataFrame(tickets2)
df2.drop_duplicates()
for index in range(0,len(df2)):
    df2.set_value(index,'Class',2)
df3 = pd.DataFrame(tickets3)
df3.drop_duplicates()
for index in range(0,len(df3)):
    df3.set_value(index,'Class',3)
df4 = df1.append(df2)
df4 = df4.append(df3)
df1[0].value_counts(sort=False).plot(kind='bar',title="First digit in 1st class` tickets numbers")
```

```python
df3[0].value_counts(sort=False).plot(kind='bar',title="First digit in 3rd class` tickets numbers")
```

# Верно ли, что женщины выживали чаще мужчин?

```python
df = pd.read_csv('train.csv')
fs = df[df['Sex']=='female']
ms = df[df['Sex']=='male']
f = [len(fs[fs['Survived']==1]),len(fs[fs['Survived']==0])]
f1 = pd.DataFrame(f)
f1.set_value(0,'S','survived')
f1.set_value(1,'S','did not survive')
m = [len(ms[ms['Survived']==1]),len(ms[ms['Survived']==0])]
m1 = pd.DataFrame(m)
m1.set_value(0,'S','survived')
m1.set_value(1,'S','did not survive')
p1 = 100*f1.iget_value(0,0)/(f1.iget_value(0,0)+f1.iget_value(1,0))
print str(p1)+'% of men survived'
p2 = 100*m1.iget_value(0,0)/(m1.iget_value(0,0)+m1.iget_value(1,0))
print str(p2)+'% of women survived'
if p1>p2:
    print 'no, men survived more often'
else:
    print 'yes, women survived more often then man'
```

# Верно ли, что чаще выживали пассажиры с более дорогими билетами?

```python
df = pd.read_csv('train.csv')
df['Survived'].corr(df['Fare'])
```

```python
df2 = df[df['Survived']==1]
print "Средняя цена билета, который 'выжил'",df2['Fare'].mean()
print "Max цена билета ",df2['Fare'].max()
print "Медиана - ",df2['Fare'].median()

df4 = df[df['Fare']>=df2['Fare'].mean()]
print "С вероятностью ",100*len(df4[df4['Survived']==1])/len(df4),"% пассажир с ценой билета более средней цены счастливого билета выживет"

df3 = df[df['Survived']==0]
print "Средняя цена билета, который 'не выжил'",df3['Fare'].mean()
print "Max цена билета ",df3['Fare'].max()
print "Медиана - ",df3['Fare'].median()

df4 = df[df['Fare']<=df3['Fare'].mean()]
print "С вероятностью ",100*len(df4[df4['Survived']==1])/len(df4),"% пассажир с ценой билета менее средней цены несчастливого билета выживет"

if df2['Fare'].mean()>df3['Fare'].mean():
    print "Да, чаще выживали пассажиры с более дорогими билетами"
else:
    print "Нет, чаще выживали пассажиры с более дешевыми билетами"
```
